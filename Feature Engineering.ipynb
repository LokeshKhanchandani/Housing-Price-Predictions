This produces lots of NaN values which might be difficult to fill and also this feature is not so profitable so better to be dropped

```python
train_df.drop('area',axis=1)
```

```python
train_df.interpolate().count()
```

```python
train_df.interpolate().plot()
```

through interpolate we will fill NaN values as they are far better than fillna() of pandas in some ways

```python
train_df.interpolate(method='akima',inplace=True)
```

‘krogh’, ‘piecewise_polynomial’, ‘spline’, ‘pchip’ and ‘akima’ are all wrappers around the scipy interpolation methods of similar names. These use the actual numerical values of the index. 

```python
train_df['bathroom'].fillna(2,inplace=True),train_df['bathroom'].fillna(2,inplace=True)
```

```python
test_df.interpolate(method='akima',inplace=True)
```

```python
test_df['location'].fillna(1,inplace=True),test_df['renovation'].fillna(1,inplace=True)
```

These are the values which remain unfilled by interpolation

```python
train_df.drop(['Garden','House ID','renovation'],axis=1,inplace=True)
test_df.drop(['Garden','House ID','renovation'],axis=1,inplace=True)
```

These features have least correlation with target so better to be dropped
