# Now lets try Heatmap for finding more correlation among features

```python
import pandas as pd
df=pd.read_csv('train_dataset.csv')
#df=df[:1650]
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks',color_codes=True)
plt.figure(figsize=(18,16))
sns.heatmap(df.astype(float).corr(),linewidths=0.1,square=True,linecolor='white',annot=True)
plt.show()
```

higher the values of corr higher the features are dependent on each other

It looks like blessing capital distance and knight house distance are closely related to each other

Also garden and kingVisit are closely related and so are curse and holy tree

```python
%matplotlib inline
df.plot(kind='scatter',x='capital_dist',y='Golden Grains')
```

Now it looks like they are 87% correlated

```python
%matplotlib inline
df.plot(kind='scatter',x='blessing',y='capital_dist')
```

They are really correlated to 100%

```python
%matplotlib inline
df.plot(kind='scatter',x='knight_house_dist',y='capital_dist')
```

```python
%matplotlib inline
df.plot(kind='scatter',x='Garden',y='kingVisit')
```

These two points shows that king only visited to those houses which have gardens
